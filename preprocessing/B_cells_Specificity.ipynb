```python
%matplotlib inline

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import os 
import numpy as np
import scipy as sci
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as pl

import scanpy as sc
import scirpy as ir
import anndata as ann

from scipy.sparse import csr_matrix
from matplotlib import rcParams
from matplotlib import colors

sc.settings.verbosity = 3
```

```python
sc.settings.set_figure_params(dpi=70)
sc.settings.verbosity = 3
sc.settings.n_jobs = 3
sc.set_figure_params(vector_friendly=True,color_map='viridis',dpi_save=600,transparent=True)
```

Paths to preprocessed data at different stages.

```python
path_base = '../data/Bcells_Covid/'
path_filtered = path_base + '01_bcrs_filtered.h5ad'
path_out = path_base + '02_bcrs_annoated.h5ad'
```

## Merge Data

```python
path_samples = f'{path_base}/raw/BCR/'
samples = list(os.listdir(path_samples))


adatas = []
for name in samples:
    print(name)
    file_name = path_base + f'raw/GEX/{name}/'
    adata_tmp = sc.read_10x_mtx(file_name)
    
    print(name)
    # TCR data
    path_anno = path_base + f'raw/BCR/{name}/filtered_contig_annotations.csv'
    anno_tcr = ir.io.read_10x_vdj(path_anno)    
    ir.pp.merge_with_ir(adata_tmp, anno_tcr)
    adatas.append(adata_tmp)
    
adata = adatas[0].concatenate(adatas[1:], batch_key='sample', batch_categories=samples)
```

```python
specificity_annotation = pd.read_csv(f'{path_base}/raw/export_probe_info copy.csv', index_col=0)
specificity_annotation.index = specificity_annotation.index.map(lambda x: x.split('_')[1] + '-' + x.split('_')[0])
specificity_annotation['Probe_binding'].value_counts()
```

```python
adata.obs['Specificity'] = specificity_annotation['Probe_binding']
adata.obs['Specificity'].value_counts()
```

## Filter based on BCR

```python
print('Cells Total: ', len(adata))
adata = adata[~adata.obs['IR_VDJ_1_cdr3'].isna()]
adata = adata[adata.obs['IR_VDJ_1_cdr3']!='None']
print('Cells with VDJ chain: ', len(adata))
adata = adata[~adata.obs['IR_VJ_1_cdr3'].isna()]
adata = adata[adata.obs['IR_VJ_1_cdr3']!='None']
print('and VJ chain: ', len(adata))
adata = adata[adata.obs['multi_chain']=='False']
print('without multi-chains: ', len(adata))
```

## Covid B cells

```python
adata = adata[adata.obs['Specificity'].isin(['Probe_Negative', 'Probe_Poly', 'Probe_ORF8', 'Probe_Spike', 
                                             'Probe_NP', 'Probe_Cocktail'])]
print('Reactive to various Covid Strains and Negative Control: ', len(adata))
```

## Quality Control

```python
adata.obs['n_counts'] = adata.X.sum(1)
adata.obs['log_counts'] = np.log10(adata.obs['n_counts'])

adata.obs['n_genes'] = (adata.X > 0).sum(1)

mt_gene_mask = [gene.startswith('MT-') for gene in adata.var_names]
mt_gene_index = np.where(mt_gene_mask)[0]
adata.obs['mt_fraction'] = adata.X[:, mt_gene_index].sum(1) / adata.X.sum(1)
```

```python
rcParams['figure.figsize'] = (15, 5)
sc.pl.violin(adata, ['log_counts'], size=1, log=False, rotation=90)
sc.pl.violin(adata, ['n_genes'], size=1, log=False, rotation=90)
sc.pl.violin(adata, ['mt_fraction'], size=1, log=False, rotation=90)
```

```python
rcParams['figure.figsize'] = (8, 8)
sc.pl.scatter(adata, y='n_genes', x='n_counts', color='mt_fraction', size=10, show=False)
```

```python
sc.pl.scatter(adata[np.logical_and(adata.obs['n_genes']<2000, adata.obs['n_counts']<10000)],
             y='n_genes', x='n_counts', color='mt_fraction', size=10, show=False)
```

## Cell Filtering

```python
print(f'Total number of cells: {adata.n_obs}')
high_mt = adata.obs['mt_fraction'] < 0.15  #todo why this value
adata = adata[high_mt]
print(f'Number of cells after MT filter: {adata.n_obs}')
```

Gene-Count Filter between 1000 and 10.000

```python
low_count = adata.obs['n_counts'] > 1500
adata = adata[low_count]
print(f'Number of cells after minimum count filter: {adata.n_obs}')

low_count = adata.obs['n_counts'] < 13000
adata = adata[low_count]
print(f'Number of cells after maximum filter: {adata.n_obs}')
```

Gene-Count Filter: filter out cells with low gene count < 500

```python
low_genes = adata.obs['n_genes'] > 500
adata = adata[low_genes]
print(f'Number of cells after minimum gene filter: {adata.n_obs}')
```

Gene-Filter: filter out genes that are represented in less than 10 cells

```python
sc.pp.filter_genes(adata, min_cells=10)
print(f'Number of Genes after gene filter: {adata.n_vars}')
```

## Simple normalization


```python
sc.pp.normalize_total(adata, target_sum=1e4)  # target sum? 
sc.pp.log1p(adata)
```

## Highly variable Gens

```python
sc.pp.highly_variable_genes(adata, n_top_genes=5000)
print('Shape before: ', adata.shape)
adata = adata[:, adata.var['highly_variable']]
print('Shape after: ', adata.shape)
```

### Clonotype Annotation

```python
adata.obs['VJ+VDJ'] = adata.obs['IR_VJ_1_cdr3'].astype(str) + '+' + adata.obs['IR_VDJ_1_cdr3'].astype(str)
clono_dict = {clone: idx for idx, clone in enumerate(adata.obs['VJ+VDJ'].unique())}
adata.obs['clonotype'] = adata.obs['VJ+VDJ'].map(clono_dict)
print(len(adata.obs['clonotype'].unique()))
adata.obs['clonotype'].head()
```
